```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img1 = np.zeros(shape=(450, 550, 3), dtype=np.uint8)

img1[100:200, 100:200, 1] = 255
img1[100:200, 100:200, 2] = 255

cv2_imshow(img1)
```

```python
img2 = np.zeros(shape=(450, 550, 3), dtype=np.uint8)
img2[150:250, 150:250, 2] = 255

cv2_imshow(img2)
```

```python
result1 = cv2.bitwise_and(img1, img2)
cv2_imshow(result1)
```

```python
result2 = cv2.bitwise_or(img1, img2)
cv2_imshow(result2)
```

```python
result3 = cv2.bitwise_xor(img1, img2)
cv2_imshow(result3)
```

```python
logo = cv2.imread("opencv_logo.png")
cv2_imshow(logo)
```

```python
logo_not = cv2.bitwise_not(logo)
cv2_imshow(logo_not)
```

## Channel Splitting

```python
logo_ch = cv2.split(logo)
for channel in logo_ch:
    print(channel.shape)
```

```python
logo_ch = cv2.split(logo)
logo_ch[2][:, :] = 0
logo_ = cv2.merge(logo_ch)
cv2_imshow(logo_)
```
