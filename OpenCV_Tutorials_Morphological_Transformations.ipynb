# Morphological Transformations
- `cv2.erode`
- `cv2.dilate`
- `cv2.morphologyEx`

## Erosion

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

im = cv2.imread('LinuxLogo.jpg', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(im, kernel, iterations=1)

cv2_imshow(im)
cv2_imshow(erosion)
```

## Dilation

```python
dilation = cv2.dilate(im, kernel, iterations=1)

cv2_imshow(dilation)
```

## Opening
Opening is the name of erosion followed by dilation.

```python
opening = cv2.morphologyEx(im, cv2.MORPH_OPEN, kernel)

cv2_imshow(opening)
```

## Closing
Closing is the name of dilation followed by erosion.

```python
closing = cv2.morphologyEx(im, cv2.MORPH_CLOSE, kernel)

cv2_imshow(closing)
```

## Morphological Gradient
It is the difference between dilation and erosion of an image.

```python
gradient = cv2.morphologyEx(im, cv2.MORPH_GRADIENT, kernel)
cv2_imshow(gradient)
```

## Top Hat
It is the difference between input image and its opening form.

```python
tophat = cv2.morphologyEx(im, cv2.MORPH_TOPHAT, kernel)

cv2_imshow(tophat)
```

## Black Hat
It is the difference between the closing of the input image and input image.

```python
blackhat = cv2.morphologyEx(im, cv2.MORPH_BLACKHAT, kernel)

cv2_imshow(blackhat)
```

## Structuring Elements

```python
rect_strel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
rect_strel
```

```python
elliptical_strel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
elliptical_strel
```

```python
cross_strel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))
cross_strel
```

```python
diamond_strel = cv2.getStructuringElement(cv2.MORPH_DIAMOND, (5, 5))
diamond_strel
```
