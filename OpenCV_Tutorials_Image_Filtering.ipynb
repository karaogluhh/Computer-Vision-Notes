```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('opencv-logo.png')


kernel = np.ones((15, 15), np.float32) /225
dst = cv2.filter2D(img, -1, kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

## `cv2.filter2D` Function Signature and Parameters

`cv2.filter2D(src, ddepth, kernel, dst=None, anchor=None, delta=0.0, borderType=None)`

**Parameters:**

*   `src`: Input image.
*   `ddepth`: Desired depth of the output image. A negative value (-1) indicates that the output image will have the same depth as the source image. For example: `cv2.CV_8U`, `cv2.CV_16S`, `cv2.CV_32F`, `cv2.CV_64F`.
*   `kernel`: Single-channel floating-point kernel matrix. For example, a 3x3 kernel: `np.ones((3,3), np.float32)/9`.
*   `dst`: Output image (optional). It should have the same size and number of channels as the input image.
*   `anchor`: Position of the anchor relative to the kernel center (optional). The default value `(-1,-1)` means that the anchor is at the kernel center.
*   `delta`: Optional value added to the filtered pixels (optional).
*   `borderType`: Pixel extrapolation method (optional). The default value is `cv2.BORDER_DEFAULT`. Other options include `cv2.BORDER_REPLICATE`, `cv2.BORDER_CONSTANT`, `cv2.BORDER_REFLECT`, `cv2.BORDER_WRAP`.

## Image Blurring
1. Blurring

```python
img = cv2.imread('opencv-logo.png')

blur = cv2.blur(img, (5,5))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

2. Gaussian Blur

```python
blur = cv2.GaussianBlur(img, (5, 5), 0)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
