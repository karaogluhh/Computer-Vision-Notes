```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("butterfly.png")
cv2_imshow(img) # cv2.imshow("Butterfly", img)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
img_flipped = cv2.imread("butterfly_flipped.png")
cv2_imshow(img_flipped) # cv2.imshow("Butterfly (flipped)", img)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
img.shape
```

```python
add_result = np.zeros(img.shape, dtype=img.dtype)
add_result = cv2.add(img, img_flipped)
cv2_imshow(add_result)
```

```python
cv2.add(img, img_flipped, add_result)
cv2_imshow(add_result)
```

```python
sub_result = np.zeros(img.shape, dtype=img.dtype)
sub_result = cv2.subtract(img, img_flipped)
cv2_imshow(sub_result)
```

```python
cv2.subtract(img, img_flipped, sub_result)
cv2_imshow(sub_result)
```
