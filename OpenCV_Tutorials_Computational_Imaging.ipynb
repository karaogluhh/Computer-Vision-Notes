```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

- `cv2.fastNlMeansDenoising`
- `cv2.fastNlMeansDenoisingColored`
- `cv2.fastNlMeansDenoisingMulti`
- `cv2.fastNlMeansDenoisingColoredMulti`


```python
im = cv2.imread("home.jpg", cv2.IMREAD_GRAYSCALE)
cv2_imshow(im)
```

```python
im[:5]
```

```python
im.shape
```

```python
noise = 10 * np.random.randn(im.shape[0], im.shape[1]).astype(np.float32)
noisy = np.clip(im.astype(np.float32) + noise, 0, 255).astype(np.uint8)
cv2_imshow(noisy)
```

```python
im_noise = im.astype(np.float32)  + noise
cv2_imshow(im_noise)
```

```python
im_noise[:5]
```

```python
denoised = cv2.fastNlMeansDenoising(noisy, None, h=12, templateWindowSize=7, searchWindowSize=21)
cv2_imshow(denoised)
```

```python
plt.subplot(121),plt.imshow(noisy)
plt.subplot(122),plt.imshow(denoised)
plt.show()
```
