```python
import cv2
from google.colab.patches import cv2_imshow
```

## Load and Display Image

The following code block loads an image named `lena.jpg` and displays it in the notebook output.

```python
img = cv2.imread("lena.jpg")

cv2_imshow(img)  #cv2.imshow("Lena", img)
cv2.waitKey(0)


cv2.destroyAllWindows()
```

## Apply Colormaps

These cells demonstrate applying different colormaps to the loaded image. The output shows the resulting image data as arrays and the visually transformed image using the SPRING colormap.

```python
cv2.applyColorMap(img, cv2.COLORMAP_JET)
```

```python
cv2.applyColorMap(img, cv2.COLORMAP_DEEPGREEN)
```

```python
img_spring = cv2.applyColorMap(img, cv2.COLORMAP_SPRING)

cv2_imshow(img_spring)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

## RGB to Gray Conversion

```python
img = cv2.imread("lena.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

## Saving an Image

```python
cv2.imwrite("lena_gray.jpg", img_gray)
```

```python
img_gray_ = cv2.imread("lena_gray.jpg")
cv2_imshow(img_gray_)
cv2.waitKey(1)
cv2.destroyAllWindows()
```
